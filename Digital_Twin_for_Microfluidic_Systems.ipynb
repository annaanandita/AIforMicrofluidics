```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_microfluidic_flow(time_steps=100):
    """Simulates fluid flow in a microfluidic channel."""
    flow = np.zeros((time_steps, 100))
    for t in range(time_steps):
        flow[t] = np.sin(np.linspace(0, np.pi, 100)) * (t / time_steps)  # Example wave
    return flow

def plot_simulation():
    """Plots a basic fluid flow simulation."""
    flow = simulate_microfluidic_flow()
    plt.imshow(flow, cmap="coolwarm", aspect="auto")
    plt.colorbar(label="Flow Velocity")
    plt.xlabel("Channel Length")
    plt.ylabel("Time Steps")
    plt.title("Microfluidic Flow Simulation")
    plt.show()

# Example Usage
plot_simulation()
```
